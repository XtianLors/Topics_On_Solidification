Because of their instability, embryos are continuously being formed and redissolved in the molten metal due to the natural agitation of atoms.

This volume free energy change, ΔGυ, is the driving force for the solidification transformation, and its magnitude is a function of temperature.

$$\Delta G^* = \frac{16\pi\gamma^3}{3(\Delta G_v)^2}$$

At the equilibrium solidification temperature $T_m$, the value of $\Delta G_v$ is zero, and with decreasing temperature its value becomes increasingly more negative. It can be shown that $\Delta G$ is a function of temperature as

$$\Delta G_v = \frac{\Delta H_f \left( T_m - T \right)}{T_m}$$

where $\Delta H_f$ is the latent heat of fusion (the heat given up during solification), and $T_m$ and the temperature T are in Kelvin.

```python
#Homoegeneous Nucleation





#Heterogeneous Nuclation

def G_crit(sigm, Tm, DH, DT):
    G_c = (((16*m.pi)/3)*(sigm**3)*(Tm**2))/((DH**2)*(DT**3))
    return G_c

def n_crit(G_c, T_t):
    n_c = spc.N_A * (np.exp(-(G_c/(spc.k*T_t))))
    return n_c 
    
```

The idealisation of solidification is stated by Stefanescu (2009, P. 50) as "either the there is enough time for both the solid and liquid solutions to become completely homogeneous, or diffusion is very rapid in both the solid and liquid."


if there is enough time for solute diffusion to proceed to completion in both the solid and the liquid the composition becomes uniform throughout the sample.This is equilibrium solificiation.


As solidification proceeds under equilibrium conditions, the solute composition in the solid, $C_s$, and in the liquid, $C_l$, vary along the solidus and the liquidus line of the phase diagram.

Knowing $C_s=kC_L$ the composition of the solid is:

$$C_s=\frac{kC_0}{1-\left( 1-k \right)f_s}$$

This equation is called the equilibrium lever rule. It is valid assuming $\rho_S=\rho_L$
